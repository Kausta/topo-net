```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from tensorflow.keras.metrics import AUC, Precision, Recall
from tensorflow.keras.optimizers import Adam
```

```python
n_classes = 2
input_dim = 200
train = pd.read_csv("train_BreastMnist224_BettiVectors_100.csv")
val = pd.read_csv("val_BreastMnist224_BettiVectors_100.csv")
test = pd.read_csv("test_BreastMnist224_BettiVectors_100.csv")
data = np.load('breastmnist_224.npz')

with open('train_breastmnist_250N.txt', 'r') as f:
    indices = [int(line.strip()) for line in f]
    
train = train.iloc[indices]
train_label = data['train_labels'][indices]
train_label = np.array([''.join(row.astype(str)) for row in train_label])

val_label = data['val_labels']
val_label = np.array([''.join(row.astype(str)) for row in val_label])

test_label = data['test_labels']
test_label = np.array([''.join(row.astype(str)) for row in test_label])

label_encoder = LabelEncoder() 
train_label = label_encoder.fit_transform(train_label) 
train_label = tf.keras.utils.to_categorical(train_label, n_classes) 

val_label = label_encoder.fit_transform(val_label) 
val_label = tf.keras.utils.to_categorical(val_label, n_classes) 

test_label = label_encoder.fit_transform(test_label) 
test_label = tf.keras.utils.to_categorical(test_label, n_classes) 
```

```python
model = Sequential([
    Dense(196, activation='relu', input_shape=(input_dim,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(n_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy', AUC(name='auc'), 'Precision', 'Recall'])
```

```python
# train model
model.fit(train, train_label, 
              epochs=100, 
              batch_size=32, 
              validation_data=(val, val_label))
test_loss, test_acc, test_auc, test_precision, test_recall = model.evaluate(test, test_label)

print(f'Test accuracy: {test_acc}')
print(f'Test AUC: {test_auc}')
print(f'Test Precision: {test_precision}')
print(f'Test Recall: {test_recall}')
```
